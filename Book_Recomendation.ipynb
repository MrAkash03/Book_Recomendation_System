```python
def check_data_info_and_shape(book, user, rating):

    info_dict = {
        'DataFrame': ['Book', 'User', 'Rating'],
        'Shape': [book.shape, user.shape, rating.shape],
        'Info': [book.info(), user.info(), rating.info()]
    }

    info_df = pd.DataFrame(info_dict)

    return info_df
```

```python
info_df = check_data_info_and_shape(book, user, rating)
info_df
```

```python
books = book.drop(['Image-URL-S','Year-Of-Publication','Image-URL-M','Image-URL-L'], axis =1)
```

```python
Book_with_ratings = rating.merge(books, on='ISBN')
```

```python
Book_with_ratings.head()
```

```python
Users_with_ratings= Book_with_ratings.merge(user, on='User-ID')
```

```python
Users_with_ratings.head()
```

```python
User_with_ratings = Users_with_ratings.drop(['Location','Age'], axis =1)
```

```python
User_with_ratings.isna().sum()
```

```python
User_with_ratings.dropna(inplace = True)
```

```python
User_with_ratings.shape
```

```python
User_with_ratings.head()
```

```python
x = User_with_ratings.groupby('User-ID').count()['Book-Rating'] > 100
most_user_atings = x[x].index
```

```python
high_ratings = User_with_ratings[User_with_ratings['User-ID'].isin(most_user_atings)]
```

```python
high_ratings.head()
```

```python
y = high_ratings.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = high_ratings[high_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.head()
```

```python
final_ratings.shape
```

```python
pivot_table = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pivot_table.fillna(0, inplace = True)
```

```python
pivot_table.head()
```

```python
pivot_table.tail()
```

```python
scaler = MinMaxScaler()
scaled_pivot = scaler.fit_transform(pivot_table)
```

```python
similarity_matrix = cosine_similarity(scaled_pivot)
```

```python
similarity_matrix.shape
```

```python
similarity_matrix
```

```python
def recommend(book_name):
    if book_name not in pivot_table.index:
        print(f"Book '{book_name}' not found in the dataset.")
        return

    index = np.where(pivot_table.index == book_name)[0][0]

    similar_items = sorted(list(enumerate(similarity_matrix[index])), key=lambda x: x[1], reverse=True)

    print(f"Recommendations for '{book_name}':")
    for i, score in similar_items[1:6]:
        book_title = pivot_table.index[i]
        author_name = book.loc[book['Book-Title'] == book_title, 'Book-Author'].values[0]
        print(f"{book_title}: {author_name}")
        print(f"Similarity Score: {score:.2f}")
        print()
```

```python
recommend('Year of Wonders')
```
